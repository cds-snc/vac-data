```python
from bs4 import BeautifulSoup
import pandas as pd
```

# Get Data

```python
df_mapping_orig = pd.read_excel('../data/VAC Data.xlsx', sheet_name=2)
```

# Munge Data

```python
df_mapping = df_mapping_orig.drop(df_mapping_orig.index[0])
df_mapping.rename(columns={
                    df_mapping.columns[0]: "result_id",
                    "Unnamed: 129": "Service Type",
                    "Unnamed: 130": "Applicant",
                    "Unnamed: 131": "Benefit"
                }, inplace=True)
df_mapping.loc[df_mapping['result_id'] == 10002, 'Benefit'] = 'Disability Pension'
```

```python
df_mapping.drop(df_mapping.columns[1:7], axis=1, inplace=True)
```

```python
df_mapping.drop(df_mapping.columns[126:], axis=1, inplace=True)
```

```python
cols = df_mapping.columns.tolist()
cols = ['Benefit', 'Applicant', 'Service Type'] + cols[:-5]
df_mapping = df_mapping[cols]
```

```python
df_mapping.drop(df_mapping.columns[1:4], axis=1, inplace=True)
identitiy_statements=['CAF - Family Member', 'Survivor', 'Orphan (Veteran is deceased)', 'War Service Veteran - Family Member', 'Allied Veteran - Family Member', 'Spouse or Common-law partner', 'Dependent Child (under the age of 18 or over 18 but unable to live independently) or between 18 and 25 and attending school', 'Caregiver', 'Dependent Child']
df_mapping.drop(identitiy_statements, axis=1, inplace=True)
```

```python
df_mapping = df_mapping.fillna(0)
df_mapping.replace(to_replace=['Y', 'N'], value=1, inplace=True)
```

# For How Many Unique Benefits is each Statement Relevant (has either a True or False for at least one of the benefits scenarios)

```python
df_mapping = df_mapping.groupby(['Benefit']).sum()
```

```python
df_mapping = df_mapping.applymap(lambda x: 1 if x > 0 else 0)
```

```python
df_result = pd.DataFrame(df_mapping.sum(axis=0))
df_result.columns = ['count']
df_result.sort_values(by=['count'], ascending=False)
```
